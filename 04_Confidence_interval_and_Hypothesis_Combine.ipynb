```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

ĐỌC DỮ LIỆU TỪ FILE CSV

```python
Data_inland = pd.read_csv('Datasets/housing_inland.csv')
Data_near_ocean = pd.read_csv('Datasets/housing_near_ocean.csv')
```

Bây giờ chúng ta sẽ kiểm định giả thuyết rằng là giá nhà trung bình của những căn nhà trong "Đất liền" và "Ven biển" có khác nhau hay không

Kiểm định giả thuyết:

+ $ H_0: \mu_1 - \mu_2 = 0 $
+ $ H_1: \mu_1 - \mu_2 < 0 $

```python
mean_inland_house_value = Data_inland['median_house_value'].describe().loc['mean']
mean_nearocean_house_value = Data_near_ocean['median_house_value'].describe().loc['mean']

std_inland_house_value = Data_inland['median_house_value'].describe().loc['std']
std_near_ocean_house_value = Data_near_ocean['median_house_value'].describe().loc['std']

n1 = Data_inland['median_house_value'].shape[0]
n2 = Data_near_ocean['median_house_value'].shape[0]
```

```python
SE = math.sqrt((std_inland_house_value**2)/n1 + (std_near_ocean_house_value**2)/n2)
Z0 = (std_inland_house_value - std_near_ocean_house_value)/SE
```

Xác định miền bác bỏ

```python
alpha = 0.05
Z_alpha = norm.ppf(1-alpha)
```

```python
p_value = norm.cdf(Z0)
if Z0 < Z_alpha: print("Bác bỏ H0")
else: print("Không bác bỏ H0")

if p_value < alpha: print("Ủng hộ H1")
else: print("Bác bỏ H1")
```
