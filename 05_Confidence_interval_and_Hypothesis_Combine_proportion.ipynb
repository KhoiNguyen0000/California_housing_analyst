```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

ĐỌC DỮ LIỆU TỪ FILE CSV

```python
Data_inland = pd.read_csv('Datasets/housing_inland.csv')
Data_near_ocean = pd.read_csv('Datasets/housing_near_ocean.csv')
```

Bây giờ chúng ta sẽ kiểm định giả thuyết rằng là tỉ lệ giá nhà trung bình > 310000 của những căn nhà trong "Đất liền" và "Ven biển" có khác nhau hay không

Kiểm định giả thuyết:

+ $ H_0: \hat{P_1} - \hat{P_2} = 0 $
+ $ H_1: \hat{P_1} - \hat{P_2} < 0 $

```python
x_inland = Data_inland[Data_inland['median_house_value'] > 310000].shape[0]
n_inland = Data_inland['median_house_value'].shape[0]

p_hat_inland = x_inland/n_inland
```

```python
x_near_ocean = Data_near_ocean[Data_near_ocean['median_house_value'] > 310000].shape[0]
n_near_ocean = Data_near_ocean['median_house_value'].shape[0]

p_hat_near_ocean = x_near_ocean/n_near_ocean
```

```python
p_hat = (x_inland + x_near_ocean) / (n_inland + n_near_ocean)
SE = math.sqrt(((p_hat) * (1 - p_hat))*((1/n_inland) + (1/n_near_ocean))) 

Z0 = (p_hat_inland - p_hat_near_ocean) / SE
```

$\alpha = 0.05$

=> $Z_{\alpha}  \approx 1.65$

```python
alpha = 0.05
Z_alpha = norm.ppf(1-alpha)
```

```python
if Z0 < Z_alpha: print("Bác bỏ H0")
else: print("Không bác bỏ H0")
```
