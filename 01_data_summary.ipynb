```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

ĐỌC DỮ LIỆU VÀ XÓA ĐI NHỮNG CỘT KHÔNG CẦN THIẾT

```python
Data = pd.read_csv('Datasets/housing.csv')
```

```python
Data_inland = Data[(Data['ocean_proximity'] == 'INLAND')] # Dữ liệu về những căn nhà trong đất liền
Data_near_ocean = Data[(Data['ocean_proximity'] == 'NEAR OCEAN')] # Dữ liệu về những căn nhà ven biển
```

```python
del_cols = ['longitude','latitude','housing_median_age',
            'total_rooms','total_bedrooms','population','households']
Data_inland = Data_inland.drop(columns=del_cols) 
Data_near_ocean = Data_near_ocean.drop(columns=del_cols)
```

VẼ BIỂU ĐỒ BOX PLOT CHO 2 DỮ LIỆU

```python
plt.boxplot([Data_near_ocean['median_house_value'],Data_inland['median_house_value']],
            patch_artist=True,
            showfliers=False,
            boxprops=dict(facecolor='lightblue',color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            flierprops=dict(marker='o',color='orange',markersize=5),
            vert=False
            )

plt.title('Box plot của giá nhà trung bình của các căn nhà trong đất liền và ven biến')
plt.yticks([1,2],['NEAR_OCEAN','INLAND'])
```

```python
plt.boxplot([Data_near_ocean['median_income'],Data_inland['median_income']],
            patch_artist=True,
            showfliers=False,
            boxprops=dict(facecolor='lightblue',color='blue'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            flierprops=dict(marker='o',color='orange',markersize=5),
            vert=False
            )

plt.title('Box plot thu nhập trung bình của chủ nhà trong đất liền và ven biến')
plt.yticks([1,2],['NEAR_OCEAN','INLAND'])
```

LƯU DỮ LIỆU LẠI THÀNH MỘT FILE CSV MỚI

```python
Data_inland.to_csv('Datasets/housing_inland.csv',index=False)
Data_near_ocean.to_csv('Datasets/housing_near_ocean.csv',index=False)
```
